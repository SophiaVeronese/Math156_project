```python
dir = '/Users/silviaveronese/Google Drive/Sophia/Spring2021/Math156/Project/Fire1/CodeAndData/fire_dataset'
dir_im_fire = dir +'/fire_images/'
dir_im_nonfire = dir +'/non_fire_images/'

lst_fire_img = glob.glob(dir_im_fire + '*.png')
lst_non_fire_img = glob.glob(dir_im_nonfire + '*.png')
```

```python
#lst_non_fire_img
print ('found ', len(lst_fire_img), ' fire images and ', len(lst_non_fire_img), 'non fire images')
```

```python
# resize images and create a new repository

# new image dimension
# lower this number if you want to run fatster

img_size_x=400
img_size_y=400

sub_fol = 'resized'+'_'+str(img_size_x)+'x'+str(img_size_y)
path_fire_resize = dir_im_fire + sub_fol 
path_non_fire_resize = dir_im_nonfire + sub_fol 


# check if dir exists if no create a new directory under each folder ( fire and non fire) to store the resized images. 
# attention: if dir exists and there are files, it will not delete old stuff. You have to do it manually. 

check_folder = os.path.isdir(path_fire_resize)

# If folder doesn't exist, then create it.
if not check_folder:
    os.makedirs(path_fire_resize)
    print("created folder : ", path_fire_resize)

else:
    print(path_fire_resize, "folder already exists.")


for i in range(len(lst_fire_img)):
    
    img = Image.open(lst_fire_img[i])
    new_img= img.resize((img_size_x,img_size_y))
    #new_name = str(img_size_x)+'x'+str(img_size_y)+lst_fire_img[i]
    new_s = '_'+str(img_size_x)+'x'+str(img_size_y)+'.png'
    new_name = lst_fire_img[i].replace('.png', new_s)
    
    new_img.save(new_name.replace('fire_images','fire_images/'+sub_fol+'/'))
    print ('creating ', new_name)


#nonfire 
    check_folder = os.path.isdir(path_non_fire_resize)

# If folder doesn't exist, then create it.
if not check_folder:
    os.makedirs(path_non_fire_resize)
    print("created folder : ", path_non_fire_resize)

else:
    print(path_non_fire_resize, "folder already exists.")


for i in range(len(lst_non_fire_img)):
    
    img = Image.open(lst_non_fire_img[i])
    new_img= img.resize((img_size_x,img_size_y))
    #new_name = str(img_size_x)+'x'+str(img_size_y)+lst_fire_img[i]
    new_s = '_'+str(img_size_x)+'x'+str(img_size_y)+'.png'
    new_name = lst_non_fire_img[i].replace('.png', new_s)
    
    new_img.save(new_name.replace('non_fire_images','non_fire_images/'+sub_fol+'/'))
    print ('creating ', new_name)
    


```
